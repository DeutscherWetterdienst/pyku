```python
cmor_variables = PYKU_RESOURCES.load_resource('cmor')
cmor_like_variables = PYKU_RESOURCES.load_resource('cmor-like')
```

```python
def _merge_variables(dict_a, dict_b):
    """
    Merge cmor_variables and cmor_like_variables dictionaries
    """
    
    duplicates = dict_a.keys() & dict_b.keys()

    if duplicates:
        raise KeyError(f"Detected duplicate keys: {duplicates}")

    return {**dict_a, **dict_b}

variables = _merge_variables(cmor_variables, cmor_like_variables)
```

```python
mapping_cds_api = PYKU_RESOURCES.load_resource('mapping_cds_api')
mapping_cordex_adjustinterp = PYKU_RESOURCES.load_resource('mapping_cordex_adjustinterp')
mapping_grib = PYKU_RESOURCES.load_resource('mapping_grib')
mapping_icon = PYKU_RESOURCES.load_resource('mapping_icon')
mapping_cordex_adjust = PYKU_RESOURCES.load_resource('mapping_cordex_adjust')
mapping_cosmo_clm = PYKU_RESOURCES.load_resource('mapping_cosmo_clm')
mapping_hostrada = PYKU_RESOURCES.load_resource('mapping_hostrada')
mapping_others = PYKU_RESOURCES.load_resource('mapping_others')
```

```python
def _add_aliases(variables, mapping):
    """
    Add aliases to variable metadata from mapping file
    """
    for key, meta in variables.items():
        entry_data = mapping.get(key, {})
        
        name_val = entry_data.get("name")
        names_val = entry_data.get("names", [])

        incoming = []
        if name_val:
            incoming.append(name_val)
        if not names_val:
            incoming.extend(names_val)

        if not incoming:
            continue
        
        aliases = meta.setdefault("aliases", [])
        for alias in incoming:
            if alias not in aliases:
                aliases.append(alias)
```

```python
_add_aliases(variables, PYKU_RESOURCES.load_resource('mapping_others'))
```

```python
mappings = [resource for resource in PYKU_RESOURCES.list_resources() if "mapping" in resource]

for mapping in mappings:
    _add_aliases(variables, PYKU_RESOURCES.load_resource(mapping))
```
